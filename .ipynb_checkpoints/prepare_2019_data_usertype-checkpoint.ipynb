```python
gender_subscriber_df = bike_df[['gender','usertype']]
```

```python
gender_subscriber_df.head(25)
```

```python
df_length = len(gender_subscriber_df)
```

```python
#divide the main_df into 1000 pieces in a list
df_increment = int(df_length / 1000)
df_increment_mod = np.mod(df_length,  1000)
print(df_increment)
print(df_increment_mod)
```

```python
gender_subscription_df_list = []

for i in range(0, 1000):
    print(gender_subscriber_df.loc[(df_increment*i):(df_increment*i + df_increment-1), :])
    inter_df = gender_subscriber_df.loc[(df_increment*i):(df_increment*i + df_increment-1), :].copy()
    for index, row in inter_df.iterrows():
        i=i+1
        #print(row)
        #print(index)
        #print('usertype', bike_df.loc[index, 'usertype'])
        #print('gender', bike_df.loc[index, 'gender'])
        if row['usertype'] == 'Subscriber':
            inter_df.loc[index, 'subscriber'] = 1
            inter_df.loc[index, 'customer'] = 0
        elif row['usertype'] == 'Customer':
            inter_df.loc[index, 'subscriber'] = 0
            inter_df.loc[index, 'customer'] = 1
        else:
            inter_df.loc[index, 'subscriber'] = 0
            inter_df.loc[index, 'customer'] = 0

        if row['gender'] == 1:
            inter_df.loc[index, 'male'] = 1
            inter_df.loc[index, 'female'] = 0
        elif row['gender'] == 2:
            inter_df.loc[index, 'female'] = 1
            inter_df.loc[index, 'male'] = 0
        else:
            inter_df.loc[index, 'female'] = 0
            inter_df.loc[index, 'male'] = 0

        if i == 10000:
            print('i=', i)
            i = 0
    gender_subscription_df_list.append(gender_subscriber_df.loc[(df_increment*i):(df_increment*i + df_increment), :])
```
