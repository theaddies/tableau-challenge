```python
# iterate over rows with iterrows()
for index, row in bike_df.head(100).iterrows():
    #print(row)
    #print(index)
    print(bike_df[index, 'usertype'])
    if row['usertype'] == 'Subscriber':
        bike_df.loc[index, 'subscriber'] = 1
        bike_df.loc[index, 'customer'] = 0
    if row['gender'] == 1:
        bike_df.loc[index, 'male'] == 1
        bike_df.loc[index, 'female'] = 0
    elif row['gender'] == 2:
        bike_df.loc[index, 'female'] = 1
        bike_df.loc[index, 'male'] == 0
    else:
        bike_df.loc[index, 'female'] = 0
        bike_df.loc[index, 'male'] == 0
```
