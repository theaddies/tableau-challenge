```python
import pandas as pd
```

```python
path = "./data/"
#old_base_file = "-citibike-stripdata.csv"
old_base_file = " - Citi Bike trip data.csv"
new_base_file = "-citibike-tripdata.csv"
year_data = "-citi-data.csv"
```

```python
#it is only the first 2 years of data that has a different name so go ahead and just rename them.

for y in range(2013, 2014, 1):
    for m in range(1, 13, 1):
        if m < 10 :
            m = '0' + str(m)
        file_name = path + str(y) + str(m) + old_base_file
        new_file_name = path + str(y) + str(m) + new_base_file
        print(file_name)
        try:
            df_opened = pd.read_csv(file_name)
            df_opened.to_csv(new_file_name, index = False)
            print("ok")
        except:
            print("error")
```

```python
#lets check the header of each file
dropped_flag = False

for y in range(2013, 2021, 1):
    for m in range(1, 13, 1):
        if m < 10 :
            m = '0' + str(m)
        new_file_name = path + str(y) + str(m) + new_base_file
        print(new_file_name)
        try:
            print('new_file_name', new_file_name)
            df_opened = pd.read_csv(new_file_name, nrows=0)
            print(df_opened)
            print("ok")

        except:
            print("error")
```

```python
#make a file of one year

for y in range(2013, 2021, 1):
    year_df = pd.DataFrame()
    for m in range(1, 13, 1):
        print('month', m)
        if m < 10 :
            m = '0' + str(m)
        new_file_name = path + str(y) + str(m) + new_base_file
        print(new_file_name)
        try:
            print('new_file_name', new_file_name)
            #index_col = 0 to not add an index, index_col = False assigns an index.
            df_opened = pd.read_csv(new_file_name, index_col = 0)
            print(df_opened)
            print("ok")
            year_df = year_df.append(df_opened)
        except:
            print("error")
    print('writing file')
    year_df.to_csv(path + str(y) + year_data, index = False) 
```
